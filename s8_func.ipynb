## Разница между return и print?

Функция, вычисляющая максимум между двумя числами, переданными в качестве аргументов:

```python
def find_max(a, b):
    if a > b:
        max_n = a
    else:
        max_n = b
    return max_n
```

```python
c = find_max(2, 1)
print(c)
```

```python
def display_max(a, b):
    if a > b:
        max_n = a
    else:
        max_ = b
    print(max_n)
```

```python
c = display_max(2, 1)
print(c)
```

Проблема с `display_max` заключается в том, что она служит только для отображения максимального числа на экране, но функция не возвращет (сохраняет) результат, а просто выводит его на экран.

## Какие ошибки обычно бывают?

1. Функция вызывается раньше, чем она была создана
2. Недостаточное количество аргументов (как при создании функции, так и при ее вызове)
3. Внутри функции использован `print` вместо `return`
4. Синтаксис (отступы, двоеточия, неправильное обращение с объектами и т.д)
5. Логические (код противоречит тому, что написано в задании)
