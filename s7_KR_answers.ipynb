### 2. Цикл while

Реализуйте программу, сохраняющую в список квадраты четных чисел:

* Задан пустой список squares. Этот код уже написан.
* В цикле while считываются целые положительные числа до тех пор, пока не введен 0 (не учитывается в последовательности).
* Для каждого вводимого числа сделайте проверку: если число четное — добавьте его в список squares.
* Гарантируется, что до 0 будет введено как минимум одно целое положительное число.
* Напечатайте список squares.

ПРИМЕР

Ввод

2 \
6 \
8 \
10 \
7 \
0

Вывод \
[4, 36, 64, 100]

```python
squares = []

# ваш код
n = int(input())
while n != 0:
    if n % 2 == 0:
        squares.append(n**2)
    n = int(input())
print(squares)
```

* В переменную years считывается список, где хранятся целые числа — год начала обучения студентов в университете. Каждый год соответствует студенту из списка students. Этот код уже написан.

* В переменную year2filter считывается число — год поступления, который нас интересует. Этот код уже написан.
* Программа выводит имена студентов, которые поступили в университет в год year2filter или позже, каждое на новой строке.
* Если таких студентов нет, то программа ничего не выводит.

ПРИМЕР

Ввод

2020, 2019, 2018 \
2019

Вывод

Петр Луков

```python
students = input().split(', ')
years = list(map(int, input().split(', ')))
year2filter = int(input())

# ваш код ниже

# 1 вариант
for student, year in zip(students, years):
    if year >= year2filter:
        print(student)
```

```python
students = input().split(', ')
years = list(map(int, input().split(', ')))
year2filter = int(input())

# ваш код ниже

# 2 вариант
for i in range(len(students)):
    if years[i] >= year2filter:
        print(students[i])
```

### 6.2 Словари

Реализуйте программу, которая проверяет, средний балл студентов.

* Дан словарь grades, в котором ключами являются имена студентов, а значениями — списки с их оценками. Гарантируется, что имена студентов не повторяются.
* В переменную threshold считывается минимально допустимый средний балл студента для прохождения конкурса на учебу по обмену. Этот код уже написан.
* Программа проверяет, что средний балл студентка не ниже значения threshold, и выводит имена таких учеников. Названия выводятся в том же порядке, в котором они хранятся в словаре. Каждое название выводится на новой строке.
* Если таких студентов нет — программа ничего не выводит.

```python
grades = {'Иван Петров': ['10', '10', '5', '6', '7', '10', '9'],
           'Петр Иванов': ['3', '4', '5', '6', '7', '10', '6'],
           'Ольга Сергеевна': ['9', '9', '8', '4', '6', '10', '10'],
           'Марина Сергеевна': ['8', '8', '8', '8', '8', '9', '9']}

threshold = int(input())

# ваш код ниже

for student, grade in grades.items():
    grade = list(map(int, grade))
    if sum(grade) / len(grade) >= threshold:
        print(student)
```

### 7. Сложный for

У Марины есть 500 рублей и список покупок. Напишите программу, которая поможет определить, на какие продукты хватит денег.

ФОРМАТ ВВОДА

* На первой строке через запятую и пробел записываются данные о товаре и о его стоимости (вида <товар>: <цена>).
* Гарантируется, что строка состоит из данных минимум об одном товаре.

ФОРМАТ ВЫВОДА

* Названия товаров, которые Марина может купить.
* Названия должны быть выведены в алфавитном порядке на одной строке через пробел.

ПРИМЕР

Ввод

хлеб: 200, яблоки: 120, молоко: 150, макароны: 300

Вывод

молоко, хлеб, яблоки

```python
# ваш код

money = 500
product_list = input().split(', ') # хлеб: 200, яблоки: 120 -> ['хлеб: 200', 'яблоки: 120']
can_buy = [] # в список добавляем продукты, которые можно купить

for n in product_list:
    product = n.split(': ')[0] # сохраняем в отдельную переменную название продукта
    price = int(n.split(': ')[1]) # сохраняем в отдельную переменную цену продукта
    if money >= price: # проверяем: если оставшихся денег больше (или равно), чем цена товара
        can_buy.append(product) # добавляем такой продукт в список
    money -= price # так как мы потратили деньги на продукт, то наши оставшиеся деньги, соответственно, уменьшаются
    
print(*sorted(can_buy), sep=', ') # сортируем и выводим список через запятую
```

### 8. While + словари

В университете проходит благотворительная лотерея. Студенты приобретают лотерейные билеты, которые вытягивают случайным образом, а через некоторое время объявляются номера выигрышных билетов. Помогите собрать информацию об участниках.

ФОРМАТ ВВОДА

* Записи о приобретенных лотерейных билетах: имя покупателя и, через точку с запятой и пробел, номер билета (целое число). Каждая запись вводится с новой строки. Имя покупателя может содержать любые символы, кроме точки с запятой.
* Один студент может купить несколько билетов, но в каждой строке указывается номер только одного билета.
* Когда записи заканчиваются, на отдельной строке вводится слово "END".

ФОРМАТ ВЫВОДА

* Словарь, где значения — имена студентов, а ключи — списки целых чисел (приобретенных лотерейных билетов).
* Ключи и значения должны быть добавлены в словарь в том же порядке, что и при вводе.

ПРИМЕР

Ввод

Harry; 20 \
Hermione; 15 \
Ron; 12 \
Harry; 21 \
Ron; 8 \
Hermione; 33 \
Hermione; 11 \
END

Вывод

`{'Harry': [20, 21], 'Hermione': [15, 33, 11], 'Ron': [12, 8]}`

```python
# вариант 1

ticket = input() 
tickets_dict = {}

while ticket != 'END':
    
    person = ticket.split('; ')[0]
    number = int(ticket.split('; ')[1])
    
    if person not in tickets_dict:
        tickets_dict[person] = []
    tickets_dict[person].append(number) 
    
print(tickets_dict)
```

```python
# вариант 2

ticket_info = input()
tickets = []
tickets_dict = {}

while ticket_info != 'END':
    tickets.append(ticket_info)
    ticket_info = input()

for ticket in tickets:
    person = ticket.split('; ')[0]
    number = int(ticket.split('; ')[1])
    if person not in tickets_dict:
        tickets_dict[person] = []
    tickets_dict[person].append(number)
    
print(tickets_dict)
```
